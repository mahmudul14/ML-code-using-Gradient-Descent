```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing files
file_path = 'Inputs_pear.xlsx'
df = pd.read_excel(file_path)
```

```python
if df.shape[1] != 11:
    raise ValueError("The input data must contain exactly 11 variables (columns)")
```

```python
correlation_matrix = df.corr(method='pearson')
print(correlation_matrix)
```

```python
# Create a heatmap
plt.figure(figsize=(10, 10))  # Increased figure size

# Create heatmap with annotation
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".4f", cmap='coolwarm', vmin=-1, vmax=1, annot_kws={"size": 8})
# Adjust the heatmap to ensure all values are visib

plt.title('Pearson Correlation Heatmap')
plt.tight_layout()  # Adjust layout to fit everything
# Save the heatmap to a file
plt.savefig('pearson_correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
```
